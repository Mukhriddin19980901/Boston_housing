```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import BatchNormalization, Conv2D, MaxPooling2D, Dense, Dropout, Flatten, Activation
from keras.callbacks import Callback
import random
from sklearn.model_selection import train_test_split
from tensorflow import keras
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.boston_housing.load_data(
    path="boston_housing.npz", test_split=0.2, seed=13)
```

```python
plt.plot(y_train,"*",color='r')
plt.ylabel("Price")
plt.xlabel("area(m**2)")
plt.show()
```

```python
model = Sequential([
    Dense(64,input_shape =[13],name='hidden_layer_1'),
    Activation('relu'),
    Dense(64,name='hidden_layer_2'),
    Activation('relu'),
    Dense(1,name='OUTPUT'),
    Activation('linear'),
],name='BostonHousingPrice')
```

```python
num_epoch=1000
model.compile(optimizer='rmsprop',loss='mse',metrics=['mae'])
hist = model.fit(x_train,y_train,epochs=num_epoch,batch_size=24)
```

```python
model.evaluate(x_test,y_test)
```

```python
preds = model.predict(x_test)
preds = preds.reshape(-1)
y_test,preds
```

```python
model.summary()
```

```python
prediction = model.predict(x_test)
plt.figure(figsize=(20,15))
plt.plot(y_test,color='r')
plt.plot(prediction,color='g')
plt.title('Housing price predictions in Boston')
plt.xlabel('area(square kilometer)')
plt.ylabel('prices')
plt.legend(['real',"prediction"])
plt.show()
```

```python
model.save('boston_housing.model',save_format='h5')
```
